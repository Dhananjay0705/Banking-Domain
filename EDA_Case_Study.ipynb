```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Banking.csv')
```

```python
## Generate descriptive statistics for the dataframe
df.describe()
```

```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Medium','High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right = False)
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

```python
# categorical columns list
categorical_cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]

# value counts
for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())

```

Univariate Analysis

```python
cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]

for i, predictor in enumerate(cols):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]

for i, predictor in enumerate(cols):
    plt.figure(i)
    sns.countplot(data=df, x=predictor,hue = 'Nationality')
```

```python
## hisplot of value counts for different occuption

for col in categorical_cols:
    if col == "Occuption" :
      continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occuption Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

Numerical Analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Amount of Credit Cards','Bank Loans','Bank Deposits','Saving Accounts','Checking Accounts','Foreign Currency Account']

## Univariate analysis and visulization
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde = True)
  plt.title(col)
plt.show()
```

Heatmaps

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Amount of Credit Cards',
                  'Bank Loans','Bank Deposits','Saving Accounts','Checking Accounts',
                  'Foreign Currency Account']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f") ## . is use for decimal
plt.title("Correlation Matrix")
plt.show()
```

**Insights of EDA :**

1. The strongest positive correlation occur among 'Bank Deposits" with "Checking Amount" , "Saving Accounts" and "Foreign Currency" indicationg that customers who mantain high balances in one account type often hold substaintial amount/funds across other accounts as well.
